```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats  as stats
from random import uniform
import pandas as pd
% matplotlib inline
```

If $y = Φ(x) = exp(x)$ and $p(x)=1$ (a uniform distribution) for $0<= x <= 1$. What is the resultant distribution for y.

Primero usaremos un método para obtener la distribución uniforme y graficaremos la distribución de frecuencias con un histograma.

```python
# Generando la distribucion uniforme para x
x_uni = np.random.uniform(0,1,10**3)

# Graficando histograma
count, bins, ignored = plt.hist(x_uni, len(x_uni))
plt.xlabel('x')
plt.ylabel('Frecuencia')
plt.title('Distribucion de frecuencias para x')
```

```python
# Definiendo la funcion y
y = np.exp(x_uni)

# Graficando histograma
count, bins, ignored = plt.hist(y, len(y))
plt.xlabel('y')
plt.ylabel('Frecuencia')
plt.title('Distribucion de frecuencias para y')
```

A continuación usaremos otro método para generar la distribución uniforme de x, usando stats.uniform.rvs.

-stats.distribution.rvs() genera numeros aleatorios de una distribución específica, en el caso de la distribución uniforme se espeficica el rango de los valores y el número de puntos a generar. 

```python
# Generando distribución uniforme
x_uniform = stats.uniform.rvs(size=10**5,loc=0,scale=1)
x_uniform.sort()
# Calculando función y con la distribución uniforme de y
y = np.exp(x_uniform)

plt.plot(x_uniform,y,label=r'f(x)=exp(x)')
plt.xlabel('x')
plt.ylabel(r'$f(x)$')
plt.legend()
```

```python
# Graficando la distribución de probabilidad para x (uniforme)
pd.DataFrame(x_uniform).plot(title='Densidad de probabilidad x',kind='density',xlim=(-0.1,1.1))
# Graficando la densidad de probabilidad de y
pd.DataFrame(y).plot(title='Densidad de probabilidad y',kind='density',xlim=(0,4))
```
