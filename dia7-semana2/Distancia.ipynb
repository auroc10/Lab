```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

# Calcular distancia entre una distribuci√≥n de puntos aleatorios 

```python
# Definimos funcion que regrese un arreglo con todas las distancias entre N puntos
def distancia(data):
    d = []
    for i in xrange(data.shape[0]-1):
        m=(((data[i+1:]-data[i])**2).sum(1))**.5
        for j in range(len(m)):
            d.append(m[j])
    return d

# Funcion que grafica el histograma del arreglo de distancias con numero de bins 'nbins'
def hist_dist(data,nbins):
    dist = distancia(data)
    count, bins, ignored = plt.hist(dist,bins = nbins,color='purple',edgecolor = 'black',  linewidth=1)
    plt.title('Histograma de distancias')
    plt.legend()
    plt.show()
    return

# Funciones que generan N puntos aleatorios uniformes en un cuadro de longitud L en 2D y 3D
def data_2D(N,L):
    data = []
    for i in range(N):
        m = [np.random.uniform()*L,np.random.uniform()*L]
        data.append(m)
    data = np.array(data)
    return data

def data_3D(N,L):
    data = []
    for i in range(N):
        m = [np.random.uniform()*L,np.random.uniform()*L,np.random.uniform()*L]
        data.append(m)
    data = np.array(data)
    return data
```

```python
# Usando un arreglo de prueba
data_test = np.array([[1,1],[2,2],[3,3],[4,4]])
hist_dist(data_test,3)
xx = [x for x,y in data_test]
yy = [y for x,y in data_test]
plt.scatter(xx,yy)
plt.title('Puntos')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Generando puntos en 2D
N = 1000 #puntos en un cuadro de distancia 1.
L = 1.
data2d = data_2D(N,L) 
hist_dist(data2d,10) 

print 'Puntos aleatorios'
# Graficando puntos puntos
xx = [x for x,y in data2d]
yy = [y for x,y in data2d]
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.scatter(xx,yy)
plt.show()
plt.clf()
```

```python
from mpl_toolkits.mplot3d import Axes3D 

# Generando puntos en 3D
N = 1000 #puntos en un cuadro de distancia 1.
L = 1.
data3d = data_3D(N,L) 
hist_dist(data2d,10) 

print 'Puntos aleatorios'
# Graficando puntos puntos
xx = [x for x,y,z in data3d]
yy = [y for x,y,z in data3d]
zz = [z for x,y,z in data3d]

fig = plt.figure()
cx = fig.add_subplot(111, projection='3d')
cx.set_aspect('equal')
cx.scatter(xx, yy, zz)
plt.show()
```
