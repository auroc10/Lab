```python
# Calcular la pendiente y punto de corte de un ajuste de minimos cuadrados

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(1,10,10)
y = [1.5,2.0,4.0,4.6,4.7,5.0,6.0,8.5,8.8,9.9]
#x_error = np.random.normal(0.1,0.2, len(elon))
#y_error = np.random.normal(0.1,30, len(masa))
```

```python
def R(x,a,b): # se define una funcion y se dice de que depende coordenada x, a, b
    return a*x+b # modelo que se calcula y= ax +b
```

```python
# minimos cuadrados con la sumatoria, notar que se resta a los datos el generado por el modelo
fx = lambda a,b:((y-R(x,a,b))**2).sum() 
fx= np.vectorize(fx)
```

```python
pend = np.linspace(-2,2,10) # definiendo un vector linea recta que será el que se explorará 14 es el numero de datos
pto_corte = np.linspace(-5,5,10) # definiendo un vector linea recta que será el que se explorará

fx_2 = fx(pend,pto_corte) # se evalua los valores creados anteriormente en la funcion chi definida

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(221) #primer plot
ax.set_xlabel(r'$a$')
ax.set_ylabel(r'$R^2$')
ax.plot(pend,fx_2,color='g')
ax.grid(True)
ax2 = fig.add_subplot(222) #segundo plot
ax2.set_xlabel(r'$b$')
ax2.set_ylabel(r'$R^2$')
ax2.plot(pto_corte,fx_2,color='y')
ax2.grid(True)

```

```python
best_values = np.where(fx_2 == np.min(fx_2))
a_best = pend[best_values]
b_best = pto_corte[best_values]
print 'The best value of the splope is', a_best
print 'the best value of intersection is', b_best
```
