```python
# Calcular la pendiente y punto de corte de un ajuste de minimos cuadrados

import numpy as np 
import matplotlib.pyplot as plt
% matplotlib inline
% config InlineBackend.figure_format = 'svg'
```

```python
# Datos a ajustar
x = np.linspace(1,10,10)
y = [1.5,2.0,4.0,4.6,4.7,5.0,6.0,8.5,8.8,9.9]
y = np.array(y)
```

```python
# Sumas necesarias para obtener la pendiente a y la interseccion b
n = len(x)
sumx = sum(x)
sumy = sum(y)
sumx2 = sum(x*x)
sumy2 = sum(y*y)
sumxy = sum(x*y)
promx = sumx/n
promy = sumy/n
```

```python
# Calculando pendiente e interseccion
a = (sumx*sumy-n*sumxy)/(sumx**2-n*sumx2)
b = promy - a*promx
```

```python
# Calculando R^2
sigmax = np.sqrt(sumx2/n - promx**2)
sigmay = np.sqrt(sumy2/n - promy**2)
sigmaxy = sumxy/n - promx*promy
R2 = (sigmaxy/(sigmax*sigmay))**2
```

```python
print 'El valor de la pendiente es ',(a)
print 'La interseccion de la abcisa es ',(b)
print 'R^2 es ',(R2)
```

```python
plt.plot(x,y,'.',label='Datos',color='purple')
plt.plot(x,a*x+b,label='Ajuste',color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste de minimos cuadrados')
plt.legend()
plt.show
```
