```python
from random import uniform
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


inside = 0 # number of point inside the circle
total = 10**3 # number of total points

# coordinates (x,y) inside the circle
x_inside = []
y_inside = []

x_outside = []
y_outside = []

for _ in range(total):
    # Generate ramdom number betwen -1 an 1 for coordinate (x,y)
    x = uniform(-1,1)
    y = uniform(-1,1)
    #condition for points inside the circle
    if np.sqrt(x**2+y**2) <= 1:
        inside += 1
        x_inside.append(x) # _.append(x) implica que se le añade a _ el valor
        y_inside.append(y)
    x = uniform(-1,1)
    y = uniform(-1,1)
    x_outside.append(x) # _.append(x) implica que se le añade a _ el valor
    y_outside.append(y)
        
        
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.scatter(x_inside, y_inside, color='red', marker='.')
ax.scatter(x_outside,y_outside,color='blue',marker='.')
plt.show
```

```python
x_inside = []
y_inside = []

x_outside = []
y_outside = []

for _ in range(total):
    # Generate ramdom number betwen -1 an 1 for coordinate (x,y)
    r = uniform(-1,1)
    theta = uniform(0,2*np.pi)
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    #condition for points inside the circle
    if (x**2+y**2) <= 1:
        inside += 1
        x_inside.append(x) # _.append(x) implica que se le añade a _ el valor
        y_inside.append(y)
    x = uniform(-1,1)
    y = uniform(-1,1)
    x_outside.append(x) # _.append(x) implica que se le añade a _ el valor
    y_outside.append(y)
        
        
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.scatter(x_inside, y_inside, color='red', marker='.')
ax.scatter(x_outside,y_outside,color='blue',marker='.')
plt.show
```
